```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_data=sc.fit_transform(data)
```

```python
X=pd.DataFrame(scaled_data)
X
```

```python
# finding wcss value for different number of clusters
from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)

  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11), wcss,'bx-')
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)
pca_data = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_data
```

```python
kmeans_model=KMeans(4)
kmeans_model.fit_predict(X)
pca_data_kmeans= pd.concat([pca_data,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_data_kmeans,palette=['red','green','blue','black'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

```python
# find all cluster centers
cluster_centers = pd.DataFrame(data=kmeans_model.cluster_centers_,columns=[data.columns])
# inverse transform the data
cluster_centers = sc.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers,columns=[data.columns])
cluster_centers
```

```python
# Creating a target column "Cluster" for storing the cluster segment
d=data
cluster_data = pd.concat([d,pd.DataFrame({'Cluster':kmeans_model.labels_})],axis=1)
cluster_data
```

```python
cluster_1_data = cluster_data[cluster_data["Cluster"]==0]
cluster_2_data = cluster_data[cluster_data["Cluster"]==1]
cluster_3_data = cluster_data[cluster_data["Cluster"]==2]
cluster_4_data = cluster_data[cluster_data["Cluster"]==3]
```

```python
#Visualization
sns.countplot(x='Cluster', data=cluster_data)
```

```python
for c in cluster_data.drop(['Cluster'],axis=1):
    grid= sns.FacetGrid(cluster_data, col='Cluster')
    grid= grid.map(plt.hist, c)
plt.show()
```

```python
cluster_data.to_csv("Clustered_Customer_Data.csv")
```

```python
#Split Dataset
from sklearn.model_selection import train_test_split
X = cluster_data.drop(['Cluster'],axis=1)
y= cluster_data[['Cluster']]
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.25)
```

```python
#Decision_Tree
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier(criterion="entropy")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
#Confusion_Matrix
from sklearn import metrics
from sklearn.metrics import classification_report
print(metrics.confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
model.score(X_test, y_test)
```
